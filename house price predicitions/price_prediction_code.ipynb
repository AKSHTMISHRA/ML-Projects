```python
# checking correlation between All the features
mp.figure(figsize=(12, 10))  # Adjust the size of the heatmap as needed
sb.heatmap(train_file.corr(), cmap='coolwarm', fmt=".2f")#annot=True,
mp.title("Correlation Heatmap")
mp.show()
```

```python
# checking correlation between salesprice and All the other features
SalePrice_corr=train_file.corr()['SalePrice']
mp.figure(figsize=(12, 10))  # Adjust the size of the heatmap as needed
sb.heatmap(SalePrice_corr.to_frame(),cmap='coolwarm', fmt='.2f')#annot=True,
mp.title("Correlation Heatmap")
mp.show()
```

From the above graph we could see some features need "label encoding",

```python
#finding all the  features for labeel encoding
all_features = train_file.columns.tolist()

# Get the features that are not present in the correlation calculation
missing_features = [feature for feature in all_features if feature not in SalePrice_corr.index]

# Print or use the missing_features list as needed
print("Features which needs label encoding")
print("-"*80)
print(missing_features)
```

```python
#label encoding 
from sklearn.preprocessing import LabelEncoder
data = train_file[['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']].copy()

# Apply label encoding to each feature in the 'data' DataFrame
label_encoder = LabelEncoder()

#applying the encoded features
# for feature in data:
#     train_file[feature] = label_encoder.fit_transform(train_file[feature])

# another method
data = data.apply(label_encoder.fit_transform)

# Replace the original columns in 'train_file' with the encoded values
train_file[['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 
           'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 
           'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 
           'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 
           'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
           'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu',
           'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 
           'SaleType', 'SaleCondition']] = data
```
