```python
#finding all the features where correlation value > 0.5

def find_high_correlation_features(train_file, target_feature, threshold=0.5):
    # Calculate the correlation matrix
    corr_matrix = train_file.corr()
    
    # Filter features with correlation greater than the threshold
    high_corr_features = corr_matrix[abs(corr_matrix[target_feature]) > threshold][target_feature]
    
    # Remove the target feature itself from the list
    high_corr_features = high_corr_features.drop(target_feature)
    
    return high_corr_features.index.tolist()


high_corr_features_gt05 = find_high_correlation_features(train_file, 'SalePrice', threshold=0.5)

print("Features with correlation > 0.5 for SalesPrice:")
print(high_corr_features)
```

```python
def find_high_correlation_features(train_file, target_feature, threshold=0.5):
    # Calculate the correlation matrix
    corr_matrix = train_file.corr()
    
    # Filter features with correlation greater than the threshold
    high_corr_features = corr_matrix[corr_matrix[target_feature] > threshold][target_feature]
    
    # Remove the target feature itself from the list
    high_corr_features = high_corr_features.drop(target_feature)
    
    # Get the correlation values for the selected features
    correlations = corr_matrix.loc[target_feature, high_corr_features.index]
    
    return high_corr_features.index.tolist(), correlations


high_corr_features, correlations = find_high_correlation_features(train_file, 'SalePrice', threshold=0.5)

print("Features with correlation value > 0.5 for SalesPrice:")
print("="*80,"\n")
for feature, correlation in zip(high_corr_features, correlations):
    print(f"{feature}: {correlation:.3f}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Assuming you have loaded your dataset into a DataFrame named 'train_file'
# And 'high_corr_features' contains the list of highly correlated feature names

# Select the independent variables (features)
x = train_file[high_corr_features]

# Select the dependent variable (target feature)
y = train_file['SalePrice']

# Split the data into training and testing sets (adjust the test_size as needed)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R2) score:", r2)
```
